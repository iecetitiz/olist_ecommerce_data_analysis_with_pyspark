### Creating Temporary SQL Views

ðŸŽ¯ Goal: Enable Spark SQL queries for Exploratory Data Analysis (EDA)

```python

# 1. Registering the Customers view
# This view represents unique customer records after the cleaning process.
customers.createOrReplaceTempView("olist_customers")

# 2. Registering the Geolocation view
# Contains latitude/longitude coordinates mapped to zip code prefixes for spatial analysis.
geolocation.createOrReplaceTempView("olist_geolocation")

# 3. Registering the Order Items view
# Details the products, prices, and freight costs associated with each individual order item.
order_items.createOrReplaceTempView("olist_order_items")

# 4. Registering the Order Payments view
# Stores payment method details and installment counts for each transaction.
order_payments.createOrReplaceTempView("olist_order_payments")

# 5. Registering the Order Reviews view
# Contains customer satisfaction scores and review comments.
order_reviews.createOrReplaceTempView("olist_order_reviews")

# 7. Registering the Products view
# Provides metadata for items, including weight, dimensions, and category names.
products.createOrReplaceTempView("olist_products")

# 8. Registering the Sellers view
# Information about the sellers who fulfilled the orders, including their location.
sellers.createOrReplaceTempView("olist_sellers")

# 9. Registering the Orders view
# The core dataset containing order status and timestamps (purchase, approval, delivery).
orders.createOrReplaceTempView("olist_orders")
```

```python
# TOTAL NUMBER OF SELLERS WHO HAVE BEEN SELLING FOR AT LEAST THREE MONTHS
```

```python
spark.sql("""
    WITH seller_stats AS (
        SELECT 
            seller_id,
            MIN(o.order_purchase_timestamp) as first_sale,
            MAX(o.order_purchase_timestamp) as last_sale
        FROM olist_order_items oi
        JOIN olist_orders o on o.order_id = oi.order_id
        WHERE o.order_status <> 'canceled'
        GROUP BY seller_id
    )
    SELECT COUNT(*) as num_of_sellers FROM seller_stats 
    WHERE MONTHS_BETWEEN(last_sale, first_sale) >= 3
""").toPandas()
```

```python
# TOTAL NUMBER OF SELLERS WHO HAVE BEEN SELLING FOR AT LEAST THREE MONTHS BY YEAR
```

```python
spark.sql("""
    WITH seller_stats AS (
        SELECT 
            seller_id,
            MIN(o.order_purchase_timestamp) as first_order,
            MAX(o.order_purchase_timestamp) as last_order
        FROM olist_order_items oi
        JOIN olist_orders o on o.order_id = oi.order_id
        WHERE o.order_status <> 'canceled'
        GROUP BY seller_id
    )
    SELECT 
        YEAR(first_order) as year,
        COUNT(DISTINCT seller_id) as num_of_sellers  
    FROM seller_stats 
    WHERE MONTHS_BETWEEN(last_order, first_order) >= 3
    GROUP BY year
    ORDER BY year
""").toPandas()
```

```python
# SALES ANALYSIS BASED ON ORDER REVIEWS/SCORES
```

```python
spark.sql("""
 SELECT 
    or_r.review_score,
    COUNT(DISTINCT o.order_id) as total_num_orders
FROM olist_order_reviews or_r
JOIN olist_orders o ON or_r.order_id = o.order_id
WHERE o.order_status <> 'canceled' AND o.order_approved_at IS NOT NULL
GROUP BY or_r.review_score
ORDER BY or_r.review_score DESC;
""").toPandas()
```

```python
# SALES ANALYSIS BASED ON ORDER REVIEWS/SCORES WITH CATEGORICAL LABELS
```

```python
spark.sql("""
    SELECT 
        CASE 
            WHEN or_r.review_score = 5 THEN 'Excellent'
            WHEN or_r.review_score = 4 THEN 'Very Good'
            WHEN or_r.review_score = 3 THEN 'Good'
            WHEN or_r.review_score = 2 THEN 'Bad'
            WHEN or_r.review_score = 1 THEN 'Very Bad'
            ELSE 'Unknown' 
        END AS review_label,
        COUNT(DISTINCT o.order_id) as total_num_orders,
        SUM(op.payment_value) as total_revenue,
        ROUND(SUM(op.payment_value) / COUNT(DISTINCT o.order_id), 2) as avg_revenue
    FROM olist_order_reviews or_r
    JOIN olist_orders o ON or_r.order_id = o.order_id
    JOIN olist_order_payments op ON op.order_id = or_r.order_id
    JOIN olist_order_items oi ON oi.order_id = or_r.order_id
    WHERE o.order_status <> 'canceled' AND o.order_approved_at IS NOT NULL
    GROUP BY review_label, or_r.review_score
    ORDER BY or_r.review_score DESC
""").toPandas()
```

```python
spark.sql("""
SELECT 
    review_score,
    COUNT(*) as total_reviews,
    SUM(CASE WHEN review_comment_message IS NULL OR review_comment_message = '' THEN 1 ELSE 0 END) as empty_reviews,
    ROUND(SUM(CASE WHEN review_comment_message IS NULL OR review_comment_message = '' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) as empty_rate_percentage
FROM olist_order_reviews
GROUP BY review_score
ORDER BY review_score DESC;
""").toPandas()
```

**High-Score Contentment:** There is a significant correlation between high satisfaction (5-4 star reviews) and blank comment fields. This suggests that satisfied customers often feel that a maximum star rating is sufficient to communicate their positive experience, resulting in a higher 'silent' majority.

**Low-Score Motivation:** Conversely, customers providing lower ratings (1 or 2 stars) show a much higher engagement in the comment section. These users are typically motivated to document specific issues, such as delivery delays or product defects.

```python
# CUSTOMER RETENTION AND LOYALTY ANALYSIS
```

```python
spark.sql("""
    WITH customer_stats AS (
        SELECT 
            c.customer_unique_id,
            COUNT(o.order_id) as num_of_orders
        FROM olist_orders o
        JOIN olist_customers c on o.customer_id = c.customer_id  
        WHERE o.order_status <> 'canceled'
        GROUP BY  c.customer_unique_id
    )
    SELECT 
        COUNT(customer_unique_id) as num_of_loyal_customers
    FROM customer_stats 
    WHERE num_of_orders > 1
""").toPandas()
```

```python
spark.sql("""
    SELECT 
        ROUND(AVG(IF(order_count > 1, 1, 0)) * 100, 2) as loyal_customer_percentage
    FROM (
        SELECT 
            c.customer_unique_id, 
            COUNT(o.order_id) as order_count
        FROM olist_orders o
        JOIN olist_customers c ON o.customer_id = c.customer_id
        WHERE o.order_status <> 'canceled'
        GROUP BY c.customer_unique_id
    )
""").toPandas()
```

ðŸ” Our analysis reveals a customer retention rate of **3.06%**, indicating that the vast majority of the Olist customer base consists of one-time buyers.

```python
spark.sql("""
    WITH customer_spending AS (
        -- Calculate total order count and total spending for each unique customer
        SELECT 
            c.customer_unique_id,
            COUNT(DISTINCT o.order_id) as order_count,
            SUM(oi.price) as total_spent
        FROM olist_orders o
        JOIN olist_order_items oi ON o.order_id = oi.order_id
        JOIN olist_customers c ON o.customer_id = c.customer_id
        WHERE o.order_status <> 'canceled'
        GROUP BY c.customer_unique_id
    ),
    revenue_metrics AS (
        -- Aggregate grand total revenue and revenue generated by loyal customers only
        SELECT 
            SUM(total_spent) as grand_total_revenue,
            SUM(IF(order_count > 1, total_spent, 0)) as loyal_customer_revenue
        FROM customer_spending
    )
    -- Final selection to calculate the revenue share percentage of loyal customers
    SELECT 
        ROUND(grand_total_revenue, 2) as total_revenue,
        ROUND(loyal_customer_revenue, 2) as loyal_revenue,
        ROUND((loyal_customer_revenue / grand_total_revenue) * 100, 2) as loyal_revenue_percentage
    FROM revenue_metrics
""").toPandas()
```

ðŸ” Loyal customers account for only **5.56%** of total revenue.

```python
spark.sql("""
    SELECT 
        oi.seller_id,
        -- Total number of unique orders per seller
        COUNT(DISTINCT o.order_id) as total_orders,
        -- Count of orders with 'canceled' status
        SUM(IF(o.order_status = 'canceled', 1, 0)) as canceled_orders,
        -- Percentage of canceled orders relative to total orders
        ROUND(AVG(IF(o.order_status = 'canceled', 1, 0)) * 100, 2) as cancellation_rate_percentage
    FROM olist_order_items oi
    JOIN olist_orders o ON oi.order_id = o.order_id
    GROUP BY oi.seller_id
    -- Sorting by total orders to see the most active sellers first
    ORDER BY total_orders DESC
    LIMIT 10
""").toPandas()
```

```python
# MOST PREFERRED PAYMENT METHOD PER PRODUCT CATEGORY
```

```python
spark.sql("""
    WITH UniqueTransactions AS (
        SELECT DISTINCT 
            p.product_category_name, 
            oi.order_id, 
            op.payment_type
        FROM olist_order_items oi
        JOIN olist_products p ON oi.product_id = p.product_id
        JOIN olist_order_payments op ON oi.order_id = op.order_id
    ),
    PaymentCounts AS (
        SELECT 
            product_category_name, 
            payment_type, 
            COUNT(*) as transaction_count
        FROM UniqueTransactions
        GROUP BY 1, 2
    ),
    RankedPayments AS (
        SELECT 
            *,
            ROW_NUMBER() OVER (PARTITION BY product_category_name ORDER BY transaction_count DESC) as rank
        FROM PaymentCounts
    )
    SELECT 
        product_category_name as category, 
        payment_type as top_payment_method, 
        transaction_count
    FROM RankedPayments
    WHERE rank = 1
    ORDER BY transaction_count DESC
""").toPandas()
```
